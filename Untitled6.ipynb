# Pandas Project

```python
import pandas as pd
```

```python
df=pd.read_csv("D:\Basic Python\Cars93.csv")
df
```

## 1.Find out cars having mileage of more than 15 miles in city?

```python
df[(df["MPG.city"]>15)] [["id","Model","Type","MPG.city"]]
```

## 2.Find out the cars with mileage in city more than 15 which are compact in type?

```python
df[(df["MPG.city"]>15) & (df.Type=="Compact")]
```

## 3.Find out Models of cars which are having Price between 17 & 25?

```python
df[(df.Price>17)&(df.Price<25)][["id","Manufacturer","Model","Price"]]
```

## 4.Find out the cars with US based origin?

```python
df[(df.Origin=="USA")][["Model","Origin"]]
```

##  5.Find out cars with no airbag and which are compact in type?

```python
df[(df.AirBags=='NaN') & (df.Type=='Compact')][['Model','MPG.city','Type','AirBags']]
```

## 6. Find out car having highest Mileage in city?

```python
df1=df.sort_values(by=['MPG.city'],ascending=False).head(1)[['id','Model','Manufacturer','MPG.city']]
df1
```

## 7. Find out top 7 most fuel economic cars for highway?

```python
df.sort_values(by=['MPG.highway'],ascending=False).iloc[0:7][['Model','Manufacturer','MPG.highway']]
```

## 8.Find out Models of cars having AirBags and which are of compact type whose price is below 20?

```python
df['AirBags']=df['AirBags'].fillna(df['AirBags'].mode().iloc[0])
```

```python
df2=df[(df.AirBags!='None') &(df.Type=='Compact') & (df.Price<20.0)][['Model','Manufacturer','AirBags','Type','Price']]
df2
```

## 9.Find out top 7 most fuel economic cars for city, show results in sorted manner based on Horsepower?

```python
df3=df.sort_values(by=['MPG.city'],ascending=False).iloc[0:7][['Model','Manufacturer','Horsepower','MPG.city']]
df3.sort_values(by=['Horsepower'])
```
